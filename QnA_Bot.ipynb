```python
from transformers import pipeline

# Load QA pipeline with distilBERT
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
context = """
The Ganges is a trans-boundary river of Asia which flows through India and Bangladesh.
The 2,525 km river originates in the western Himalayas and flows southeast through the Gangetic Plain.
It is the most sacred river to Hindus and is worshipped as the goddess Ganga.
"""
```

```python
question = "Define Ganga"

result = qa_pipeline(question=question, context=context)
print(f"Answer: {result['answer']}")
```
